```python
def multiply_brute_force(a,n):
    result = 0
    
    for i in range(0,n):
        result = result+a
    return result

def multiply_divide_conquer(a,n):
    if (n == 0):
        return 0
    elif (n == 1):
        return a
    elif (n % 2 == 0):
        half_product = multiply_divide_conquer(a, n / 2)
        return half_product + half_product
    else:
        half_product = multiply_divide_conquer(a, (n - 1) / 2)
        return (a + half_product + half_product)
    

def main():

    a=int(input("Enter 'a': "))
    n=int(input("Enter 'n': "))
    print("Brute Force:")
    print(multiply_brute_force(a, n))
    print("Divide and Conquer:")
    print(multiply_divide_conquer(a, n))
if __name__=="__main__":
    main()  
```

```python
import time
import random
import matplotlib.pyplot as ply
def main():
    no=int(input("Enter the number of time sort is to be implemented: "))
    N=[]
    time=[]
    A=[] 
    for i in range(0,no):
        n=int(input(f"Enter the no. of elements for sort {i+1}: "))
        for j in range(0,n):
            A.append(random.randrange(0,1000))
        time.append(quickSort(A,0,n-1))
        N.append(n)
    ply.plot(time,N)
    ply.xlabel("time(seconds)")
    ply.ylabel("Number of elements(int)")
    ply.show()   
    
def partition(a,low,high):
    p=a[high]
    i=low-1
    for j in range(low,high):
        if a[j]<=p:
            i=i+1
            (a[i],a[j])=(a[j],a[i])
    (a[i+1],a[high])=(a[high],a[i+1])
    return i+1
def quickSort(array, low, high):
    start=time.time()
    if low < high:
        pi = partition(array, low, high)
        quickSort(array, low, pi - 1)
        quickSort(array, pi + 1, high)
    end=time.time()
    return(end-start)
if __name__=="__main__":
    main()  
```

```python
def binomialCoefficient_brute_force(n,k):
    return (factorial(n)//(factorial(k)*factorial(n-k)))
def factorial(n):
    t=1
    for i in range (1,n+1):
        t=int(t*i)
    return int(t)
def binomialCoefficient_dynamic(n,k):
    if k > n:
        return 0
    dp = [[0] * (k + 1) for _ in range(n + 1)]

    for i in range(n + 1):
        dp[i][0] = 1

    for i in range(1, n + 1):
        for j in range(1, min(i, k) + 1):
            dp[i][j] = dp[i - 1][j - 1] + dp[i - 1][j]

    return dp[n][k]


def main():
    print("Enter 'n' & 'k' where n > k")
    n=int(input("Enter 'n': "))
    k=int(input("Enter 'k': "))
    print("Brute Force:")
    print(binomialCoefficient_brute_force(n, k))
    print("Dynamic: ")
    print(binomialCoefficient_dynamic(n, k));
if __name__=="__main__":
    main()  
```

```python
import math
def brute_poly(poly,n,x):
    result=0
    l=n-1
    for i in range (0,n):
        result=result+poly[i]*math.pow(x,l)
        l=l-1
    return int(result)
def horner_poly(poly,n,x):
    result=poly[0]
    for i in range(1,n):
        result=result*x+poly[i]
    return result
def main():
    n=int(input("Enter the size of the polynomial (Number of coefficients in the polynomial): "))
    print("Enter the polynomial coefficients: ")
    poly=[]
    for i in range (0,n):
        a=int(input("Enter: "))
        poly.append(a)
    x=int(input("Enter the value of 'x': "))
    print("Brute Force Algoritm: ")
    print(brute_poly(poly,n,x))
    print("Horner's Method: ")
    print(horner_poly(poly,n,x))
if __name__=="__main__":
    main()
```

```python
def BoyerMoore(pattern, text):
    m = len(pattern)
    n = len(text)
    if m > n: 
        return -1
    table = []
    for k in range(256): 
        table.append(m)
    for k in range(m - 1): 
        table[ord(pattern[k])] = m - (k + 1)
    table = tuple(table)
    k = m - 1
    while k < n:
        j = m - 1; 
        i = k
        while j >= 0 and text[i] == pattern[j]:
            j -= 1; 
            i -= 1
        if j == -1: 
            return i + 1
        k += table[ord(text[k])]
    return -1
def main():
    text=input("Enter the text: ")
    pattern=input("Enter the pattern: ")
    s = BoyerMoore(pattern, text)
    print ('Text:',text)
    print ('Pattern:',pattern)
    if s > -1:
        print ('Pattern \"' + pattern + '\" found at position',s)
if __name__ == '__main__':
    main();
```

```python
def KMP_String(pattern, text):
    a = len(text)
    b = len(pattern)
    prefix_arr = get_prefix_arr(pattern, b)
    initial_point = []
    m = 0
    n = 0
    while m != a:
        if text[m] == pattern[n]:
            m +=  1
            n += 1
        else:
            n = prefix_arr[n - 1]
        if n == b:
            initial_point.append(m - n)
            n = prefix_arr[n - 1]
        elif n == 0:
            m += 1
    return initial_point
def get_prefix_arr(pattern, b):
    prefix_arr = [0] * b
    m = 1
    n = 0
    while m != b:
        if pattern[m] == pattern[n]:
            n += 1
            prefix_arr[m] = n
            m += 1
        elif n != 0:
            n = prefix_arr[n - 1]
        else:
            prefix_arr[m] = 0
            m += 1
    return prefix_arr
def main():
    string = input("Enter the String: ")
    pat = input("Enter the pattern: ")
    initial_index = KMP_String(pat, string)
    print("String :",string)
    print("Pattern :",pat)
    for i in initial_index:
        print('Pattern is found in the string at index number', i)
if __name__=="__main__":
    main()
```

```python
visited=[]
queue=[]
def BFS(node,root):
    visited.append(root)
    queue.append(root)
    while queue:
        m=queue.pop(0)
        print(m,end=" ")
        for n in node[m]:
            if n not in visited:
                visited.append(n)
                queue.append(n)
def main():
    n=int(input("Enter the no. of nodes: "))
    dict_node={}
    for i in range (0 , n):
        node=int(input("Enter the node: "))
        neigh=int(input(f"Enter the no. of neighbour {node} has: "))
        x=[]
        for j in range (0,neigh):
            a=int(input(f"Enter neighbour no. {j+1}: "))
            x.append(a)
        dict_node.update({node:x})
    root=int(input("Enter the root node: "))
    if root not in dict_node.keys():
        print("Invalid Root")
    else:
        print("BFS: ")
        BFS(dict_node,root)
if __name__=="__main__":
    main()
```

```python
import itertools
def sub_set(S,n,d):
    count=0
    for i in range(0,n+1):
        for j in itertools.combinations(S,i):
            if sum(j)==d:
                print(list(j))
                count+=1
    if count==0:
        print("Matching Subset not Found")
def main():
    n=int(input("Enter the size of the Set: "))
    S=[]
    print("Enter the Elements: ")
    for i in range (0,n):
        a=int(input("Enter: "))
        S.append(a)
    print(S)
    d=int(input("Enter the Sum 'd': "))
    sub_set(S,n,d)
if __name__=="__main__":
    main()
```
